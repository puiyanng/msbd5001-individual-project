```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
trainDataframe = pd.read_csv("train.csv")
testDataframe = pd.read_csv("test.csv")

concat = trainDataframe.append(testDataframe, ignore_index=True, sort=True)
concat.fillna(0,inplace=True)
concat.drop([ "purchase_date", "release_date","is_free","tag"], axis = 1, inplace=True)
```

```python
#temp = pd.get_dummies(pd.concat([trainDataframe,testDataframe],keys=[0,1]), columns=['x'])

# Selecting data from multi index and assigning them i.e
#Xtrain,Xtest = temp.xs(0),temp.xs(1)


#encoder = OneHotEncoder(categories = "genres")
#X_train_encoded = encoder.fit_transform("X_train")
#X_test_encoded = encoder.transform("X_test")

x = concat.categories.str.split(',',expand=True).stack().reset_index(level=1, drop=True).to_frame('category') 
x = pd.get_dummies(x, prefix='c', columns=['category']).groupby(level=0).sum()
x
concat = pd.concat([concat, x], axis = 1) #default ignore_index is False

concat.drop(["categories"], axis = 1, inplace=True)
concat
```

```python
x = concat.genres.str.split(',',expand=True).stack().reset_index(level=1, drop=True).to_frame('genre') 
x = pd.get_dummies(x, prefix='g', columns=['genre']).groupby(level=0).sum()

concat = pd.concat([concat, x], axis = 1) #default ignore_index is False

concat.drop(["genres"], axis = 1, inplace=True)
```
